```python
# ein Tuple erstellen 

my_tuple = ("apple", "banana", "cherry")
print(my_tuple)
```

```python
#ein Tuple ist indiziert

my_tuple = ("apple", "banana", "cherry")
print(my_tuple[0])
print(my_tuple[1])
print(my_tuple[2])
```

```python
# zwei werte sind erlaubt

my_tuple = ("apple", "banana", "cherry", "cherry")
print(my_tuple)
```

```python
# mit len()-Methode bekommen die Anzahl der Elemente des Tuples aus

my_tuple = ("apple", "banana", "cherry")
print(len(my_tuple))
```
