```python
#hier aktualisieren wir den Tuple, und fügen wir eine neue element ein!

my_tuple = ("apple", "banana", "cherry")
new_list = list(my_tuple)               #machen den Tuple eine Liste
new_list.append("orange")               #fügen wir eine neue element mit append()-methode ein
my_tuple = tuple(new_list)              #aktualisieren der Tuple mit die neue elemente

print(my_tuple)
```
