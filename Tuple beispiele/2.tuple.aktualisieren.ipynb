```python
#hier aktualisieren wir den Tuple, und fügen wir eine neue element ein!

my_tuple = ("apple", "banana", "cherry")
new_list = list(my_tuple)               #machen den Tuple eine Liste
new_list.append("orange")               #fügen wir eine neue element mit append()-methode ein
my_tuple = tuple(new_list)              #aktualisieren der Tuple mit die neue elemente

print(my_tuple)
```

```python
#so können wir auch eine element in einem Tuple hinzufügen

my_tuple1 = ("element1", "element2")
my_tuple2 = ("element3",)               #hier ist wichtig komma zeichen dazu schreibe, weil ohne ist ein String

my_tuple1 = my_tuple1 + my_tuple2

print(my_tuple1)
```

```python
# element einer Tuple entfernen(geht nicht), aber es gibt ein weg

my_tuple = ("apple", "banana", "cherry")
my_list = list(my_tuple)        #erstellen eine liste mit den elementen von tuple
my_list.remove("apple")         #entferne wir das Element von der Liste
my_tuple = tuple(my_list)       #aktualisieren wir die tuple liste

print(my_tuple)
```

```python
#tuple löschen

my_tuple = ("apple", "banana", "cherry")
print(my_tuple)
del my_tuple
print(my_tuple)             #error weil es keine tupel existiert
```
