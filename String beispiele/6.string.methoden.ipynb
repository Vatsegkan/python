```python
# capitalize() wandelt das erste Zeichen in Großbuchstaben um
#nimmt keine parameter
#wenn die erste Zeichen Zahl ist, wird nicht passiert!

text = "wir werden die Methode capitalize verwenden"
text1 = "python ist FUN"
new_text = text.capitalize()
new_text1 = text1.capitalize()      #hier wird nur die erste Buchstabe groß und die restliche klein!
print(new_text)
print(new_text1)
```

```python
# find() und index() durchsucht die zeichenfolge nach einem bestimmten Wert und gibt die Position zurück
#an der er gefunden wurde
#wenn nicht finden gibt -1 zurück   ->  gibt error zurück
#Syntax     ->      string.find/index(value, start, end)

text = "hier probieren wir der find methode"
new_text = text.find("probieren")
print(new_text)

new_text = text.find("nicht")       #Hier bekommen wir -1, da wir kein Wert gefunden haben!
print(new_text)

new_text = text.find("e")
print(new_text)                     #hier gibt die erste zeichen in zeichenfolge^

new_text = text.find("e", 5, 11)
print(new_text)                     #wo im text der Buchstabe "e" zum erste mmal vor, wenn wir nur zwischen
                                    #Position 5 und 10 suchen


print(text.find("g"))           #hier gibt -1 zurück
print(text.index("g"))          #hier gibt Error zurück
```
