```python
#capitalize wandelt das erste Zeichen in Großbuchstaben um
#nimmt keine parameter
#wenn die erste Zeichen Zahl ist, wird nicht passiert!

text = "wir werden die Methode capitalize verwenden"
text1 = "python ist FUN"
new_text = text.capitalize()
new_text1 = text1.capitalize()      #hier wird nur die erste Buchstabe groß und die restliche klein!
print(new_text)
print(new_text1)
```
