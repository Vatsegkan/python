```python
# Großbuchstaben

#1-te weg

text = "wir werden mit Großbuchstaben schreiben!"
print(text.upper())

#2-te weg
new_text = "wir werden mit Großbuchstaben schreiben!".upper()
print(new_text)
```

```python
#Kleinbuchstaben

#1-te weg

text = "wir werden mit Großbuchstaben schreiben!"
print(text.lower())

#2-te weg
new_text = "wir werden mit Großbuchstaben schreiben!".lower()
print(new_text)
```

```python
#Leerzeichen entfernen
#1-te Weg

text = "    wir werden die Leerzeichen entfernen.    "
print(text.strip())

#2-te Weg
new_text = "    wir werden die Leerzeichen entfernen.    ".strip()
print(new_text)
```

```python
#Zeichen ersetzen

text = "wir werden ein String mit einem anderen string ersetzen"
print(text.replace("e", "o"))       #mit replace()-Methode erzetzen wir einen string mit einem anderen string
                                    #da wo 'w' sind werden jetze 'e' sein
```

```python
#split()-Methode gibt eine Liste zurück

probe1 = "wir, probieren, die, split, methode"
print(probe1.split(","))       #hier sagen wir ->  ignoriere diese Zeichen "," und gib die listen zurück

probe2 = "wir. probieren die. split. methode"
print(probe2.split("."))        #hier sagen wir ->  ignoriere diese Zeichen "" und gib die listen zurück

probe3 = "wir probieren die split methode"
print(probe3.split())           #falls wir keine Zeichen haben, lassen wir die Methodeparameter leer!
```
