```python
#Slicing, man kann hier ein String über index schneiden!!

text = "hier werden wir den String schneiden, einfach so!!!"

print(text[:36])        #hier haben wir keinen anfangswert, also fängt von index 0 ein!
print(text[0:5])        # Hier fängt von 0 bis 5-1
print(text[38:])        # Hier haben wir anfangswert aber kein endwert, also fängt von index 38 bis ende
print(text[:])          # so geben wir das ganze text
print(text[-13:-1])     # hier haben wir negation, fängt von ganz recht mit -1 ... 
                        # da wir zwei werte eingegeben haben, gilt am ende n-1
```
