```python
#string sind arrays

text = "Hier haben wir einen Array"
print(text[0])          #über eckige klammern greifen wir auf den array zu!
```

```python
#Da Strings Array sind, können wir die Zeichen in einem string mit einer for schleife durchlaufen

for buchstabe in "bannana":
    print(buchstabe)        #hier werden die buchstaben einzeln ausgegeben!
```

```python
#String prüfen

text = "wir werden nach einem Wort suchen!"

activ = True

count = 0
while activ:
    count += 1
    if count == 10:
        print("Sie haben keine versuche mehr!")
    
    wort = input("Geben Sie bitte eine Wort ein:")
    if wort in text:                                            #hier kann man auch not in verwenden
        print("Wir haben gefunden, das Wort ist: "+ wort)       
        print("Sie hatten "+str(count)+" versuche gehabt!")
        break
    else:
        print("versuchen Sie noch man!")
        continue
```
