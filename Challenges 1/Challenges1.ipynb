```python
tagen = int(input("Geben Sie der Anzahl an Tagen ein:"))

stunden = tagen * 24
minuten = stunden * 60
sekunden = minuten * 60

print("Tagen:"+str(tagen))
print("Stunden:"+str(stunden))
print("Minuten:"+str(minuten))
print("Sekunden:"+str(sekunden))
```
