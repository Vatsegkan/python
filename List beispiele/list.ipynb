```python
#Auf Listenelemente zugreifen

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

#zugreifen mit dem Index, hier ist wichtig zu beachten, wenn wir auf anfangswert und ende wert zugreifen möchte
#ende wert ist immer n-1

print(my_list[2:5])         #hier lauft von index 2 bis index 4 (NICHT BIS INDEX 5!!!)

print(my_list[:4])          #da wir kein IndexAnfangsWert haben, fängt automatisch von index 0 bis index 3

print(my_list[2:])          #da wir kein IndexEndeWert haben, fängt von index 2 und gibt alle elemente
                            # der Liste aus (Hier gilt nicht n-1)

print(my_list[-4:-1])       #Negation (rechtes nach links), der letze Element der liste ist -1
                            #hier wird also von orange bis melon laufen (hier gilt n-1)

if "apple" in my_list:      #mit der in-operator Prüft man, ob ein element in der liste ist oder nicht
    print("passt")
else:
    print("blabla")
```

```python
#Listenelemente hinzufügen

my_list = ["apple", "banana", "cherry"]
my_list.append("orange")                    # mit der .append() können wir am ende der liste einen Element 
print(my_list)                              #hinzufügen

my_list.insert(1, "orange")                 # mit der insert() können wir auf ein index ein element hinzufügen
print(my_list)                              #und die alte elemente der liste nach dem aktuelen index nach rechts verschieben


my_tropical_list = ["mango","anannas","papaya"] #mit .extend() andere Liste am ende der aktuelle liste hinzufügen
my_list.extend(my_tropical_list)                
print(my_list)


my_tuple_list = ("kiwi", "cherry")          #mit .extend() können wir jedes iteriebares Objekt
my_list.extend(my_tuple_list)               #z.B tuples, sets, dictionaries etc hinzufügen
print(my_list)
```

```python
#Listenelemente entfernen

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

my_list.remove("apple")             #remove() entfernt das angegebene Element
print(my_list)

my_list.pop(0)                      #pop() entfernt den angegebenen Index
print(my_list)

my_list.pop()                       #wenn wir kein Index eigeben, löscht pop()-Methode dsa letzte element
print(my_list)  

del my_list[0]                      #del entfernt auch das angegebenen Index ...
print(my_list)

#del my_list                         #kann auch die Liste auch komplett löschen!

my_list.clear()                      #clear macht die Liste leer!
print(my_list)
```

```python
#Liste kopieren

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]


new_list = my_list.copy()
print(new_list)
```

```python
#die Reihefolge der Liste ändern

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(my_list)
my_list.reverse()
print(my_list)
```

```python
#Schleife durch Liste

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
for element in my_list:                 
    print(element)                      #wir können die Listenelemente mit einer for schleife durchlaufen
                                        #hie drücken wir alle elemente einzeln
```

```python
#... wir können die Listenelemente auch durchlaufen, indem wir ihre Indexnummer verweisen

my_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

for index in range(len(my_list)):       #length ist hier 7 elemente, das heißt range ist von 0 bis 6
    print(my_list[index])
```
