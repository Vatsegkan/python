```python
#C08:

#Hier entfernen wir die Zeichen "c"
alphabet = ["a", "b", "c", "c", "d", "e", "e"]
alphabet.remove("c")
print(alphabet)

#Hier sagen wir, wenn "f nicht in alphabet liste ist, dann weise "f" auf die letzt stelle der Liste auf!
if "f" not in alphabet:
    alphabet[-1] = "f"
print(alphabet)


#Hier weisen wir die Elemente der Liste in den Variablen. Man kann das auch mit einem Schritt machen
variableIndex1 = alphabet[0]
variableIndex2 = alphabet[1]
#variableIndex1, variableIndex2 = alphabet[:2]            #mit einem Schritt

print(variableIndex1,variableIndex2)
```
