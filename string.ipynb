```python
#string Anführungszeichen

print("Hallo")          #die beide Beispiele haben das gleiche Bedeutung
print('hallo')          #...und machen das gleiche arbeit
```

```python
#zuweisung einer String in einer Variable

wort = "Hallo"
print(wort)
```

```python
#mehrzeilige Zeichenfolgen

text = """wir schreiben
in drei Zeile
und weisen wir das ganze in einer variable zu"""    # das funktioniert auch mit ''' ... '''
print(text)
```

```python
#string sind arrays

text = "Hier haben wir einen Array"
print(text[0])          #über eckige klammern greifen wir auf den array zu!
```

```python
#Da Strings Array sind, können wir die Zeichen in einem string mit einer for schleife durchlaufen

for buchstabe in "bannana":
    print(buchstabe)        #hier werden die buchstaben einzeln ausgegeben!
```
